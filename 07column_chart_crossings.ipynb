## Çaprazlamalar

```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
cut_kategoriler = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
```

```python
sns.catplot(x = "cut", y = "price", data = df);
```

```python
#barplotu fonksiyonunu kullandığımızda hue ile 3. boyutu eklediğimizde oluşan grafikte y ekseni bozulmuş gibi gözükse de
#aslında boyutlar arasındaki değişkeni ifade etmek için temsil ifadesi yapılmış demektir.
```

```python
sns.barplot(x = "cut", y = "price", hue = "color", data =df );
```
